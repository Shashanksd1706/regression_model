```python
df=df[['Country','EdLevel','YearsCodePro','Employment','ConvertedCompYearly']]
```

```python
df.rename(columns={'ConvertedCompYearly':'Salary'},inplace=True)
df.rename(columns={'YearsCodePro':'Experience'},inplace=True)
```

```python
df.head(50)
```

```python
df=df[df['Salary'].notnull()]  ## getting those records whose salary is not null
```

```python
df
```

```python
df.info()
```

```python
#checking for the null values in dataset
df.isnull().sum()
```

```python
df=df.dropna()  ## dropping the null values from the dataset
df.isnull().sum()
```

```python
df=df[df['Employment']=='Employed, full-time']
df=df.drop(columns=['Employment'])
df.info()  ## dropping the employment column as we have filtered the data for full time employees
```

```python
df
```

```python
df['Country'].value_counts()
```

```python
def shorten_categories(categories,cutoff):
    categorial_map={}
    for i in range (len(categories)):
        if categories.values[i]>=cutoff:
            categorial_map[categories.index[i]]=categories.index[i]
        else:
            categorial_map[categories.index[i]]='Other'
    return categorial_map
```

```python
country_map=shorten_categories(df['Country'].value_counts(),500)
df['Country']=df['Country'].map(country_map)
df['Country'].value_counts()
```

```python
%matplotlib inline
fig,ax=plt.subplots(1,1,figsize=(15,10))
df.boxplot('Salary','Country',ax=ax)
plt.suptitle('Salary VS Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Experience'].unique()
```

```python
def clean_experience(x):
    if x=='Less than 1 year':
        return 0.5
    if x=='More than 50 years' :
        return 50
    return float(x)
df['Experience']=df['Experience'].apply(clean_experience)    
```

```python
df['EdLevel'].unique()
```

```python
def clean_education(x):
    if 'Master’s degree ' in x:
        return 'Masters degree'
    if 'Bachelor’s degree'in x:
        return 'Bachelors Degree'
    if 'Professional degree' in x or 'Doctoral Degree' in x:
        return 'Post Grad'
    return 'Less than a Bachelors'
df['EdLevel']=df['EdLevel'].apply(clean_education)                
```

```python
df['EdLevel'].unique()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_education=LabelEncoder()
df['EdLevel']=le_education.fit_transform(df['EdLevel'])
df['EdLevel'].unique()
```

```python
df
```

```python
le_country=LabelEncoder()
df['Country']=le_country.fit_transform(df['Country'])
df['Country'].unique()
```

```python
df
```

USING MACHINE LEARNING TECHNIQUE

```python
x=df.drop('Salary',axis=1)
y=df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
linear_reg=LinearRegression()
linear_reg.fit(x,y.values)
```

```python
y_pred=linear_reg.predict(x)
```

PERFORMENCE METRICES

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
error=np.sqrt(mean_squared_error(y,y_pred))
```

```python
print("${:,.2f}".format(error))
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg=DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(x,y.values)
```

```python
y_pred=dec_tree_reg.predict(x)
```

```python
error=np.sqrt(mean_squared_error(y,y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_for_reg=RandomForestRegressor(random_state=0)
random_for_reg.fit(x,y.values)
```

```python
y_pred=random_for_reg.predict(x)
```

```python
error=np.sqrt(mean_squared_error(y,y_pred))
print("${:,.2f}".format(error))
```

```python
from sklearn.model_selection import GridSearchCV

max_depth=[None,2,4,6,8,10,12]
parameters={'max_depth':max_depth}

regressor=DecisionTreeRegressor(random_state=0)
gs=GridSearchCV(regressor,parameters,scoring='neg_mean_squared_error')
gs.fit(x,y.values)
```

```python
regressor=gs.best_estimator_

regressor.fit(x,y.values)
y_pred=regressor.predict(x)
error=np.sqrt(mean_squared_error(y,y_pred))
print("${:,.2f}".format(error))
```
